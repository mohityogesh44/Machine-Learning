```python
import pandas as pd
import numpy as np
```

```python
names = ['id', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion',
         'Single Epithelial Cell', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data',names = names)
df.head()
```

```python
df.drop(columns = ['id'], inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.replace(to_replace='?',value = np.nan, inplace = True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,9), dpi = 100)
sns.heatmap(df.corr(), annot = True, cmap = 'viridis')
```

```python
sns.pairplot(df, diag_kind= 'kde')
```

```python
x = df.drop('Class', axis = 1)
x
```

```python
y = df.Class
y
```

```python
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.svm import SVC
```

#### Linear Kernel

```python
#LINEAR KERNEL
model = SVC(kernel='linear')
model.fit(x_train, y_train)
```

```python
preds = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
accuracy_score(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```

#### Polynomial Kernel

```python
model2 = SVC(kernel = 'poly')
model2.fit(x_train, y_train)
```

```python
preds2 = model2.predict(x_test)
```

```python
print(accuracy_score(y_test, preds2))
```

```python
print(classification_report(y_test, preds2))
```
